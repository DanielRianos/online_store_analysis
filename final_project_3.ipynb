# Products Analysis

```python
import pandas as pd
```

## Combining Dataframe with all the products information

### Data set 7: Olist Products

```python
data7_products= pd.read_csv('brazilian_e_commerce_public/data_sets/7_olist_products_dataset.csv')
data7_products.head()
```

### Data set 9: Product category name translation

```python
data9_translation= pd.read_csv('brazilian_e_commerce_public/data_sets/9_product_category_name_translation.csv')
data9_translation.head()
```

```python
products_list= data7_products.merge(data9_translation, how='left',on= 'product_category_name')
```

```python
products_list.head()
```

```python
products_list.info()
```

```python
data_f1_olist = pd.read_csv('final_datasets/data_f1_olist.csv')
data_f1_olist.head()
```

```python
data_f1_olist.info()
```

```python
data_f1_olist_with_products= data_f1_olist.merge(products_list, how='left',on= 'product_id')
```

```python
data_f1_olist_with_products.head()
```

```python
data_f1_olist_with_products.info()
```

```python
data_f1_olist_with_products['order_id'].nunique()
```

```python
data_f1_olist_with_products['product_category_name_english'].nunique()
```

```python
data_f1_olist_with_products['sub_group_rank']= data_f1_olist_with_products.groupby('distance_group')['delta_purchase_delivered_customer'].rank(ascending=False, method = 'first').astype(int)
```

```python
data_f1_olist_with_products.head()
```

## Creating Dataframe for the top 200 products with the highest product journey per distance group

```python
df_top_200 = data_f1_olist_with_products[top_200]
df_top_200
```
