# Averages per distance groups calculations

```python
import pandas as pd
```

```python
data_f1_olist = pd.read_csv('final_datasets/data_f1_olist.csv')
data_f1_olist.head()
```

```python
data_f1_olist.count()
```

```python
data_f1_olist.info()
```

### Average order price per distance group

```python
average_price_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['price'].mean()),2)
average_price_per_distance_group
```

### Average freight value per distance group

```python
average_freight_value_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['freight_value'].mean()),2)
average_freight_value_per_distance_group
```

### Average time between the purchase and the order approvement per distance group

```python
average_delta_purchase_approved_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['delta_purchase_approved'].mean()),2)
average_delta_purchase_approved_per_distance_group
```

### Average time between the purchase and the delivery to the carrier per distance group

```python
average_delta_purchase_delivered_carrier_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['delta_purchase_delivered_carrier'].mean()),2)
average_delta_purchase_delivered_carrier_per_distance_group
```

### Average product journey  per distance- Average time between the purchase and the delivery to the customer per distance group

```python
average_delta_purchase_delivered_customer_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['delta_purchase_delivered_customer'].mean()),2)
average_delta_purchase_delivered_customer_per_distance_group
```

### Average time between the carrier and the delivery to the customer per distance group

```python
average_delta_delivered_customer_carrier_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['delta_delivered_customer_carrier'].mean()),2)
average_delta_delivered_customer_carrier_per_distance_group 
```

### Average time between delivery to the customer and the estimated delivery time per distance group

```python
average_delta_delivered_customer_order_estimated_delivery_per_distance_group = round(pd.DataFrame(data_f1_olist.groupby('distance_group')['delta_delivered_customer_order_estimated_delivery_'].mean()),2)
average_delta_delivered_customer_order_estimated_delivery_per_distance_group
```
